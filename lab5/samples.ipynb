```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x0 = cv2.imread('samples/x0.png')
x1 = cv2.imread('samples/x1.png')
x01 = np.array([x0, x1]) 
```

```python
plt.imshow(x0)
```

```python
idx = np.array([
    [0, 1],
    [1, 0],
    [1, 1]
])
x = np.zeros((90, 60, 3), dtype=int)
for j in range(0, 3):
    x[j*30:(j+1)*30, :] = np.concatenate(x01[idx[j]], axis=1)

plt.imshow(x)
cv2.imwrite('samples/sample_1.png', x)
```
